```python
import pandas as pd
import seaborn as sns
```

```python
body_df = pd.read_csv('data/body.csv', index_col = 0)
body_df.head()
```

```python
body_df['Height']
```

```python
body_df['Height'].value_counts().sort_index()  # index 순으로 정렬
```

```python
body_df['Height'].value_counts().sort_index().plot()
```

```python
sns.kdeplot(body_df['Height'], bw_method=0.2)
```

## 히스토그램

```python
body_df.plot(kind='hist', y='Height', bins=15)
```

```python
sns.distplot(body_df['Height'], bins=15)
```

```python
body_df.plot(kind='box', y='Height')  # 자료 전체에 대한 요약
```

```python
sns.violinplot(y=body_df['Height'])  # 분포 전체를 보여줌
```

```python
body_df.plot(kind='scatter', x='Height', y='Weight')
```

```python
sns.kdeplot(body_df['Height'], body_df['Weight'])  # 등고선
```
